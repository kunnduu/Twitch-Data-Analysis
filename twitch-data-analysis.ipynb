## 1. % of streamers being partnered with twitch and the % of streamers producing mature content

```python
colors = [ '#B7C3F3','#4F6272']
colors1= ['#DD7596', '#8EB897']
plt.subplot(1, 2, 1)
raw_df.groupby(by=["Partnered"])['Partnered'].count().plot(kind='pie',figsize=(15,10),autopct='%1.2f%%',fontsize=15, wedgeprops = { 'linewidth' : 4, 'edgecolor' : 'black' },colors=colors)
plt.title("% of streamers being twitch partners",fontsize=15)
plt.subplot(1, 2, 2)
plt.title("% of streamers producing or having there content marked as mature",fontsize=15)
raw_df.groupby(by=["Mature"])['Mature'].count().plot(kind='pie',figsize=(15,10),autopct='%1.2f%%',fontsize=15, wedgeprops = { 'linewidth' : 4, 'edgecolor' : 'black' },colors=colors1)
plt.show()
```

* We can see that 97.80% that is 978 streamers are partnered with twitch and the rest 2.20% that is 22 stramers are not 
* We can see that 77% of streamers that is 770 streamers do not make mature content where as the the rest 230 does make mature content

## 2. Finding out the composition of the language in which most of the streamers stream

```python
Lang=raw_df.groupby(by=["Language"], dropna=False).count() #First groupping the dataframe according to the language
```

```python
Lang
```

```python
channel=(Lang.Channel) #From the above table we can see that due grouping we can find the number of channels using a particular language
channel
```

```python
channel=dict(channel) #Creating a dictionoary
channel={k: v for k, v in sorted(channel.items(), key=lambda item: item[1])} #and sorting it according to the value or the number channels
```

```python
channel
```

```python
#Seperating the languages and the number of channels using the language
channellist=(channel.values())
languagelist=(channel.keys())
```

```python
import plotly.express as px
fig = px.funnel_area(names=languagelist,
                    values=channellist,
                    title='<span style="font-size: 20px;">Top languages in which streamers stream(interactive)</span>')
fig.show()
```

From above funnel chart representation we can clearly see that the highest used language is English that is 48.5% and the second and third most used language to stream is Korean and Russian

## 3. Different distributions

```python
plt.figure(figsize=(20,10))
sns.set(style="darkgrid")
fig = sns.kdeplot(raw_df['Followers'], shade=True, color="r");
fig.set_xlabel("Density of followers",fontsize=25)
fig.set_ylabel("Density",fontsize=25)
fig.set_title("Distribution of followers",fontsize=30)
plt.show()
```

We can see that the followers have a high density between 0 and 0.2x10^7 followers
so we can say that we are observing the data about some of the top streamers.

```python
plt.figure(figsize=(20,10))
sns.set(style="darkgrid")
fig = sns.kdeplot(raw_df['Average viewers'], shade=True, color="g");
fig.set_xlabel("Average viwers",fontsize=25)
fig.set_ylabel("Density",fontsize=25)
fig.set_title("Distribution of Average viewers",fontsize=30)
plt.show()
```

Has a high density between 0-20000 average viewers.

```python
import plotly.express as px
fig = px.violin(raw_df, y="Peak viewers", color="Partnered", box=True, points="all",
          hover_data=raw_df.columns,title='<span style="font-size: 20px;">Peak viewers density according to partnership(interactive)</span>')
fig.show()
```

* Partnered streamers have a high density of peak viewership around 16.5k
* Non partnered streamers have a high density of peak viewership around 21.3k 

## 4. Top 10 streamers according to there Average Viewers

```python
df=raw_df.sort_values(by='Average viewers' ,ascending=False).head(10)
df
```

```python
def addlabels(x,y):
    for i in range(len(x)):
        plt.text(i,y[i],y[i])
```

```python
plt_1 = plt.figure(figsize=(20, 10))
sns.set_theme(style="darkgrid")
plt.xticks(rotation=45)
sns.barplot(x="Channel", y="Average viewers" ,data=df,palette='GnBu')
addlabels((df['Channel']).tolist(), (df['Average viewers']).tolist())
plt.xlabel("Name of the streamer/channel",fontsize=20)
plt.ylabel("Average viewers ",fontsize=20)
plt.title("Top 10 streamers according to the Average viewers",fontsize=25)
plt.show()
```

 ## 5. Top 10 Streamers based on Followers

```python
df=raw_df.sort_values(by='Followers' ,ascending=False).head(10)
plt_1 = plt.figure(figsize=(20, 10))
sns.set_theme(style="darkgrid")
plt.xticks(rotation=45)
sns.barplot(x="Channel", y="Followers" ,data=df,palette='YlOrRd')
addlabels((df['Channel']).tolist(), (df['Followers']).tolist())
plt.xlabel("Name of the streamer/channel",fontsize=20)
plt.ylabel("Followers",fontsize=20)
plt.title("Top 10 streamers according to Followers",fontsize=25)
plt.show()
```

## Q1 How much does the viewers gained and the followers gained is effected by being a twitch partner or not

```python
f, axes = plt.subplots(1, 2)
from matplotlib import rcParams

# figure size in inches
rcParams['figure.figsize'] = 11.7,8.27
sns.boxplot(x='Partnered',y='Followers gained',palette="Set1",data=raw_df,ax=axes[0])
sns.boxplot(x='Partnered',y='Views gained',palette="Set2",data=raw_df,ax=axes[1])
plt.show()
```

* We can see that the **followers gained** and the **views gained are more** when the streamers are **partnered** with twitch
* This maybe due to **twitch pushing** the **content** for the streamers **partnered with them**.
* We can also say that the **followers gained** and the **views gained** are kind of **interlinked more the followers more the number of views.**

## Q2 Which language have the highest views gained by the streamers

```python
langview=raw_df.groupby(by=["Language"], dropna=False).sum()
langview=langview.drop('Other').sort_values(by='Views gained',ascending=False)
langview
```

```python
plt_1 = plt.figure(figsize=(20, 10))
sns.set_theme(style="darkgrid")
plt.xticks(rotation=45)
sns.barplot(x=langview.index, y="Views gained" ,data=langview,palette='cool')
addlabels((langview.index).tolist(), (langview['Views gained']).tolist())
plt.xlabel("Language",fontsize=20)
plt.ylabel("Total viewers ",fontsize=20)
plt.title("Total viewers gained by a language",fontsize=25)
plt.show()
```

* From the above plot we can see that the language which gained the **most number of views** is **English** and **after that** is **Russian**.
* The reason maybe **English being a global language** many **countries teach the language in there schools** increasing the viewership around the world

## Q3 Effect of stream time and watch time on followers gained

```python
df=raw_df[raw_df['Followers gained']>=0]
```

```python
import plotly.express as px
fig = px.scatter(df, x="Stream time(minutes)", y="Watch time(Minutes)", size='Followers gained',color='Mature',title='<span style="font-size: 20px;">Peak viewers density according to partnership(interactive)</span>')
fig.show()
```

* A very different thing comming up in this as the size of the bubble represents the Followers gained the most followers gained are for low streaming time and high watch time 
* Maybe this shows us the ones being present on the platform for a longer time does not increase your chance of gaining the followers but the type of content bring catered to the viewers is affecting it more 
* Which is less mature content which helps us reach all the type of viewers
